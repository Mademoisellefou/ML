```python
import  numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
def gradiente_descendiente(x,y,rt):
    b=0
    m=0
    _range=1000
    lr=rt
    n=len(x)
    for i in range (_range):
        y_pred=(m*x)+b
        mse=(1/n)*sum([2**val for val in (y-y_pred)])
        dm=(2/n)*sum(-x*y+x*y_pred)
        db=(2/n)*sum(-y+y_pred)
        m = m - lr*dm
        b = b - lr*db
        print('m {} ,b {},cost  {},i {}'.format(m,b,mse,i))
```

```python
gradiente_descendiente(x,y,0.025)
```

```python
def gradiente_descendiente1(x,y,rt):
    b=0
    m=0
    mse_before=0
    _range=1000
    lr=rt
    n=len(x)
    for i in range (_range):
        y_pred=(m*x)+b
        mse=(1/n)*sum([2**val for val in (y-y_pred)])
        dm=(2/n)*sum(-x*y+x*y_pred)
        db=(2/n)*sum(-y+y_pred)
        m = m - lr*dm
        b = b - lr*db
        if math.isclose(mse,mse_before,rel_tol=1e-05):
            return m,b
        mse_before=mse
        print('m  {} ,  b {} ,cost  {},i {}'.format(m,b,mse,i))
    return m,b
```

```python
m,b=gradiente_descendiente1(x,y,0.025)
```

```python
print('Corelation {} , interception {}'.format(m,b) )
```
